```python
#lets move towards models
X = mpg_car_dumied.drop('mpg',axis=1)
y = mpg_car_dumied['mpg']
```

```python
#Split data train and test

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=30,random_state=1)
```

```python
#Invoking the linear regression model from sklearn

from sklearn.linear_model import LinearRegression
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train,y_train)
```

```python
#Evaluating the model
#score is a gives - R2 or coefficient of determinant
#R2 = 1 - RSS /TSS


linear_regression_model.score(X_test,y_test)
```
