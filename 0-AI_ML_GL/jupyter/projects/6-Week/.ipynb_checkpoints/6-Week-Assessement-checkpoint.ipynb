## 3. Drop columns: “date”, “volume” and “symbol”

# 4 . Get “close” as the target column and rest of the columns as features 

```python
X = stock_trimmed.loc[:,stock_trimmed.columns != 'close']
y = stock_trimmed['close']
X.head()
```

# 9. Define the sequential model 

```python
nn_reg_model = Sequential()
```

## 9.a	Add a flatten layer 

```python
nn_reg_model.add(Flatten(name="Initial_Flatten_Layer",input_shape=[3]))
```

## 9.b  Add Dense layer with linear activation function and one neuron

```python
nn_reg_model.add(Dense(units=1,activation='linear',name="Linear_Activated_Dense_Layer",))
```

# 10. Compile the model
10.a Use loss as “mean squared error”
10.b Use optimizer as “sgd”

```python
def coeff_determination(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred )) 
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) ) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

<font color='green'> 
****
## Evaluating Regression

1)As we are performing regression Accuracy is not a correct measure for evaluation.
    
2) **R-Squared** should be used to measure how close the predected values are to the original. Here we have defined R-Square  in a function called coeff-determination
    
3) We use **MSE** and **MAE** to have a measure of the loss. Which are Mean Square Error and Mean Absolute Error


****
</font>
