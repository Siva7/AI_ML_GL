```python
import warnings
warnings.simplefilter('ignore')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

from statsmodels.stats.weightstats import ttest_ind
```

```python
#Importing Data
data_df = pd.read_csv(r"statsmodels-building-statistical-models\Datasets\jobtest.csv")
data_df
```

```python
data_df.boxplot(column=['JPERF'],by='MINORITY')
```

```python
minority_df = data_df.loc[data_df['MINORITY'] == 1,:]
non_minority_df = data_df.loc[data_df['MINORITY'] == 0,:]
minority_df
```

```python
ttest_ind(minority_df['JPERF'],non_minority_df['JPERF'],alternative='two-sided',value=0)
```

```python
#T test : Comparing between people who socre 
#         Above the median and below

low_test_scores = data_df.loc[data_df['TEST'] < data_df['TEST'].median(),:]
high_test_scores = data_df.loc[data_df['TEST'] >= data_df['TEST'].median(),: ]


low_test_scores
```

```python
ttest_ind(high_test_scores['JPERF'],low_test_scores['JPERF'],alternative='two-sided',value=0)
```
