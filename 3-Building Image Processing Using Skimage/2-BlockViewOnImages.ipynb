```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import data
%matplotlib inline

from skimage import color
from skimage.util import view_as_blocks
import skimage.io
```

```python
three_dogs = skimage.io.imread('images/pexels-3-dogs.jpg')
print(three_dogs.shape)
print(three_dogs.ndim)
three_dogs_gray = color.rgb2gray(three_dogs)
print(three_dogs_gray.shape)
print(three_dogs_gray.ndim)
plt.imshow(three_dogs_gray,cmap='gray')
```

```python
block_shape=(4,4)
```

```python
three_dogs_gray_block = view_as_blocks(three_dogs_gray,block_shape=block_shape)
```

```python
print(three_dogs_gray_block.shape)
print(three_dogs_gray_block.ndim)
three_dogs_gray_block

c = np.array([[1],[1,2]])
print(c.ndim)
print(c.shape)
```

```python
flatten_blocks = three_dogs_gray_block.reshape(three_dogs_gray_block.shape[0],three_dogs_gray_block.shape[1],-1)
print(flatten_blocks.shape)
print(flatten_blocks.ndim)
```

```python
mean_blocks = np.mean(flatten_blocks,axis=2)
mean_blocks.shape
```

```python
plt.imshow(mean_blocks,cmap='gray')
```

```python
plt.imshow(mean_blocks,cmap='gray',interpolation='nearest')
```

```python
max_blocks = np.max(flatten_blocks,axis=2)
max_blocks.shape
plt.imshow(max_blocks,cmap='gray',interpolation='nearest')
```

```python
min_blocks = np.min(flatten_blocks,axis=2)
min_blocks.shape
plt.imshow(min_blocks,cmap='gray',interpolation='nearest')
```
