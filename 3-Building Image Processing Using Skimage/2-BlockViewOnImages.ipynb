```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import data
%matplotlib inline

from skimage import color
from skimage.util import view_as_blocks
import skimage.io
```

```python
three_dogs = skimage.io.imread('images/pexels-3-dogs.jpg')
three_dogs_gray = color.rgb2gray(three_dogs)

plt.imshow(three_dogs_gray,cmap='gray')
```

```python
block_shape=(4,4)
```

```python
three_dogs_gray_block = view_as_blocks(three_dogs_gray,block_shape=block_shape)
```

```python
three_dogs_gray_block.shape
```

```python
flatten_blocks = three_dogs_gray_block.reshape(three_dogs_gray_block.shape[0],three_dogs_gray_block.shape[1],-1)
flatten_blocks.shape
```
