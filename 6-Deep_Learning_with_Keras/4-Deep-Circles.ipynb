```python
from sklearn.datasets import make_circles
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
```

```python
def plot_data(pl,X,y):
    pl.plot(X[y==0,0],X[y==0,1],'ob',alpha=0.5)
    pl.plot(X[y==1,0],X[y==1,1],'xr',alpha=0.5)
    pl.legend(['0','1'])
    return pl
```

```python
def plot_decision_boundary(model, X, y):

    amin, bmin = X.min(axis=0) - 0.1
    amax, bmax = X.max(axis=0) + 0.1
    hticks = np.linspace(amin, amax, 101)
    vticks = np.linspace(bmin, bmax, 101)
    
    aa, bb = np.meshgrid(hticks, vticks)
    ab = np.c_[aa.ravel(), bb.ravel()]
    
    # make prediction with the model and reshape the output so contourf can plot it
    c = model.predict(ab)
    Z = c.reshape(aa.shape)

    plt.figure(figsize=(12, 8))
    # plot the contour
    plt.contourf(aa, bb, Z, cmap='bwr', alpha=0.2)
    # plot the moons of data
    plot_data(plt, X, y)

    return plt
```

```python
X,y = make_circles(n_samples=1000,factor=.6,noise=0.1,random_state=1)
```

```python
p1 = plot_data(plt,X,y)
p1.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import plot_model
```

```python
model = Sequential()
```

```python
model.add(Dense(units=1,activation='sigmoid',input_shape=(2,)))
model.compile(optimizer=Adam(lr=0.05),loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
eval_result = model.evaluate(X_test, y_test)
```

```python
plot_decision_boundary(model, X, y).show()
```

## Trying for a better model

```python
model2 = Sequential()
step1 = Dense(units=4,activation='tanh',input_shape=(2,),name="initial_Layer")
step2 = Dense(units=4,activation='tanh',name="Second_Layer")
step3 = Dense(units=1,activation='sigmoid',name="final_layer")
model2.add(step1)
model2.add(step2)
model2.add(step3)
model2.compile(optimizer=Adam(lr=0.05),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model2.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
plot_decision_boundary(model2, X, y).show()
```

# Adding Call backs to the training

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
my_calback = [EarlyStopping(monitor='val_accuracy',patience=5,min_delta=1)]
```

```python
model2.fit(X_train,y_train,epochs=100,verbose=1,callbacks=my_calback,validation_data=(X_test,y_test))
```
